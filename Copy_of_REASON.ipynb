```python
import os
import numpy as np
import pickle
from sentence_transformers import SentenceTransformer

# Define the Google Drive path where files are stored
drive_path = '/content/drive/My Drive/image_project'

# Load image embeddings
image_embeddings = np.load(os.path.join(drive_path, 'image_embeddings.npy'))

# Load description embeddings
description_embeddings = np.load(os.path.join(drive_path, 'description_embeddings.npy'))

# Load image paths
with open(os.path.join(drive_path, 'image_paths.pkl'), 'rb') as f:
    image_paths = pickle.load(f)

# Load the text embedding model
text_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
import pandas as pd
import torch
from torchvision import models, transforms
from PIL import Image
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to get the closest image based on a query question
def get_closest_image(query, threshold=0.5):
    # Get text embedding for the query
    query_embedding = text_model.encode([query])[0].reshape(1, -1)

    # Compute cosine similarities between the query and the image descriptions
    description_similarities = cosine_similarity(query_embedding, description_embeddings)

    # Find the index of the most similar description
    most_similar_idx = description_similarities.argmax()
    similarity_score = description_similarities[0, most_similar_idx]

    # Check if similarity score meets the threshold
    if similarity_score < threshold:
        return "No relevant image found for this query.", None

    # Retrieve the corresponding image path and similarity score
    closest_image_path = image_paths[most_similar_idx]
    return closest_image_path, similarity_score
```
